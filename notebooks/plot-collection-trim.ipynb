```python
plt.rc('font', size=22, weight='bold', family='sans-serif')
plt.rc('axes', titlesize=22)#, labelsize=18)
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)
plt.rc('legend', fontsize=22)
plt.rc('figure', titlesize=22)
```

```python
name = "HPK"
vendor = "Vendor 1"

#name = "QPT"
#vendor = "Vendor 2"

outdir = name+"-plots"
```

```python
# Pixel plots
allpix = pd.read_csv(name+"-plots/allpix.csv")
allpix_bad = pd.read_csv(name+"-plots/allpix_bad.csv")

npix = len(allpix["pa"])
print(npix)

logmax = npix*50
```

```python
def draw_plot(x,y,xbounds,ybounds):

    xarray = np.clip(allpix[x],xbounds[0],xbounds[1])
    yarray = np.clip(allpix[y],ybounds[0],ybounds[1])

    plt.hist2d(x=xarray,y=yarray,bins=[np.linspace(xbounds[0],xbounds[1],60),np.linspace(ybounds[0],ybounds[1],60)],norm=matplotlib.colors.LogNorm())
    plt.xlabel(x,fontweight='bold')
    plt.ylabel(y,fontweight='bold')

    plt.xlim(xbounds)
    plt.ylim(ybounds)
    cbar = plt.colorbar()
    cbar.set_label("Pixels ("+name+")",fontweight='bold')
    
    plt.tight_layout()
    
```

```python
x="trimbits"; y="THR_Mean"

xbounds = [-20,60]
ybounds = [0,160]

draw_plot(x,y,xbounds,ybounds)
plt.plot([-30,70],[85,85],color='black',linestyle='--')

plt.savefig(outdir+"/hist2d_"+x+"_"+y+".pdf")
plt.savefig(outdir+"/hist2d_"+x+"_"+y+".png")
```
